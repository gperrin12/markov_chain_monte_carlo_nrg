## Define a log likelihood function, and set priors.

```python
# what do I know about my model? e.g. the slope should be between 0 and 90 degree
def lnprior(theta):
    m1, m2, b = theta
    if (-10 < m1 < 10) and (-10 < m2 < 10) and (0 < b < 10):
        return 0.0
    return -np.inf
```

```python
#remember the chi square: 
# Sum((model - data)/error)^2 is a quantity that follows a chi square distribution with N=number of degreed of freedom. 
# Divide it by the number of degreed  of freedom as Chisq = (Sum((model - data)/error)^2)/DOF
# If it is ~1 that is a good model. 
# >1 bad model
# <1 your errors must be underestimated (model too good to be true!)

def lnlike(theta, x, y, yerr):
    m1, m2, b = theta
    model = b + m1*x + m2*(x**2)
    dof = len(theta)
    #return -(1.0 / dof) * np.sum(((model - y) / yerr)**2)
    return -(1.0) * (np.sum((model-y)**2 / yerr**2))
```

```python
# ENCEE needs a lnprob function to put together prior and posterior (in log space)
def lnprob(theta, x, y, yerr):
    lp = lnprior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(theta, x, y, yerr)
```

## set up the MCMC and run it

```python
np.random.seed(123)

# Choose the "true" parameters.
m1_true = -1.170558
m2_true = 0.427849
b_true = 7.645410

theta0 = (m1_true, m2_true, b_true)

# we start the walk here
walk = [theta0]

next_step = (m1_true, m2_true, b_true) + np.random.randn(3)

print (theta0, next_step)
```

```python
x = bblnrgdataCut.UnitsTotallog
y = bblnrgdataCut.nrglog
yerr = bblnrgdataCut.errorlog
```

```python
theta_current = (m1_true, m2_true, b_true) + np.random.randn(3)

# we start the walk here
walk = [theta_current]

# propoed perturbation of the current position - gaussian with width 0.5
proposal_width = (0.2, 0.2, 0.2) 
proposal = spnorm(theta_current, proposal_width).rvs()

# current values  of the pieces of Bayes theorem
ln_current = lnlike(theta_current, x, y, yerr)
prior_current = lnprior(theta_current)

# values  of the pieces of Bayes theorem at the proposed position
ln_proposal  = lnlike(proposal, x, y, yerr)
prior_proposal = lnprior(proposal)

# put them together
p_current = np.exp(ln_current + prior_current)
p_proposal = np.exp(ln_proposal + prior_proposal)
```

```python
N = 50000 #number of steps in my chain

# move along the walk N = 100, 1000 steps
np.random.seed(666)

theta_current = (m1_true, m2_true, b_true) + np.random.randn(3)

# we start the walk here
walk = [theta_current]

# proposed perturbation of the current position - gaussian with width 0.5
proposal_width = (0.5, 0.5, 0.5) 
proposal = spnorm(theta_current, proposal_width).rvs()
#print proposal
# current values  of the pieces of Bayes theorem
ln_current = lnlike(theta_current, x, y, yerr)
prior_current = lnprior(theta_current)
#print ln_current
# values  of the pieces of Bayes theorem at the proposed position
ln_proposal  = lnlike(proposal, x, y, yerr)
prior_proposal = lnprior(proposal)
#print ln_proposal

# put them together
p_current = ln_current + prior_current
p_proposal = ln_proposal + prior_proposal

for i in range(N):  
    # calculate the proposal and the prior and likelihood there
    # update for sure if it is a better location
    if p_proposal > p_current:
        print ("right away")
        theta_current, ln_current, prior_current, p_current = \
        proposal, ln_proposal, prior_proposal, p_proposal
        walk.append(proposal)      
    #update with a probability equal to the posteriors ratio if the location is worse
    else:
        p_accept = (p_proposal / p_current)     
        #use a stocastic draw to accept with probability P
        if np.random.rand() < p_accept:
            #print ("later on away")
            theta_current, ln_current, prior_current, p_current = proposal, ln_proposal, prior_proposal, p_proposal
            walk.append(proposal)
    proposal = spnorm(theta_current, proposal_width).rvs()
    ln_proposal  = lnlike(proposal, x, y, yerr)
    prior_proposal = lnprior(proposal)
    #p_proposal = np.exp(ln_proposal + prior_proposal)
            
m_mcmc = np.percentile(np.array(walk)[:,0], [50, 16, 84])
b_mcmc = np.percentile(np.array(walk)[:,1], [50, 16, 84])
mcmcRes = np.concatenate([m_mcmc, b_mcmc])
print ("MCMC result:\nslope\t\t{0:+.2f} ({1:.2f}, {2:.2f})\nintercept\t{3:.2f} ({4:.2f}, {5:.2f})".format(*mcmcRes))
```

```python
# look at the sampled posterior  space
pl.figure(figsize=(10,5))
pl.plot(np.array(walk)[:,0],'k-', label="slope m1")
pl.legend()
pl.show()
print ("Fig 1: m1 slope as a function of step number in markov chain")

pl.figure(figsize=(10,5))
pl.plot(np.array(walk)[:,1], 'k-', label="slope m2")
pl.legend()
pl.show()
print ("Fig 2: m1 slope as a function of step number in markov chain")

pl.figure(figsize=(10,5))
pl.plot(np.array(walk)[:,2], 'k-', label="intercept")
pl.legend()
pl.show()
print ("Fig 3: intercept as a function of step number in markov chain")

pl.figure()
pl.hist(np.array(walk)[:,0], alpha=0.5, label="slope m1")

pl.hist(np.array(walk)[:,1], alpha=0.5, label="slope m2")

pl.hist(np.array(walk)[:,2], alpha=0.5, label="intercept")
pl.legend()
pl.show()
print ("Fig 4:  slopes and intercept MCMC posterior distributions")
```

As you can see in Figures 1, 2, and 3, even at step 50,000, the parameters do not converge to a single value, meaning that either the number of steps is insufficient, or more walkers need to be added to converge. I'll add more walkers by trying to determine the parameters with the emcee library below.

```python
ndim, nwalkers, nsteps = 3, 100, 500
pos = [np.array([m1_true, m2_true, b_true]) + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=(x, y, yerr))
sampler.run_mcmc(pos, nsteps);
```

```python
for i in range(sampler.chain.shape[-1]):
    ax = plt.figure().add_subplot(1,1,1)
    for j in range(sampler.chain.shape[0]):
        ax.plot(range(sampler.chain.shape[1]), sampler.chain[j,:,i], 'k', alpha=0.1)
```

As you can see in the figures above, the "burn in" point for each parameter is right around 150 steps, so looking at the chain after 150 steps should give the converged parameter solutions

```python
samples = sampler.chain[:, 150:, :].reshape((-1, ndim))
fig = corner.corner(samples, labels=["$m1$","$m2$", "$b$"], truths=(m1_true, m2_true, b_true))
pl.show()
print ("Fig 6: EMCEE with {0:d} chains of {1:d} steps  ".format(nwalkers, nsteps) + 
       "- generative model truth indicated in blue")
```

Yes! The parameters m1, m2, and b converge to tight surfaces in the plots above, signifying that those parameters and the quadratic model chosen fit the data very well.

```python
x = bblnrgdataCut[['UnitsTotallog', 'NumFloors_log', 'RetailArea', 'ResArea', 'BldgArea_log']]
y = bblnrgdataCut.nrglog
yerr = bblnrgdataCut.errorlog
```

```python
# what do I know about my model? e.g. the slope should be between 0 and 90 degree
def lnprior(theta):
    m1, m2, m3, m4, m5, b = theta
    if (-10 < m1 < 10) and (-10 < m2 < 10) and (0 < b < 10)\
    and (-10 < m3 < 10) and (-10 < m4 < 10) and (-10 < m5 < 10):
        return 0.0
    return -np.inf
```

```python
#remember the chi square: 
# Sum((model - data)/error)^2 is a quantity that follows a chi square distribution with N=number of degreed of freedom. 
# Divide it by the number of degreed  of freedom as Chisq = (Sum((model - data)/error)^2)/DOF
# If it is ~1 that is a good model. 
# >1 bad model
# <1 your errors must be underestimated (model too good to be true!)

def lnlike(theta, x, y, yerr):
    m1, m2, m3, m4, m5, b = theta
    model = b + m1*x.iloc[:,0] + m2*x.iloc[:,1] + m3*x.iloc[:,2] + m4*x.iloc[:,3] + m5*x.iloc[:,4]
    dof = len(theta)
    #return -(1.0 / dof) * np.sum(((model - y) / yerr)**2)
    return -(1.0) * (np.sum((model-y)**2 / yerr**2))
```

```python
### look at autocorrelation
def acorr(x):
    ax = pl.figure(figsize=(10,5)).add_subplot(111)

    x = x - x.mean()

    autocorr = np.correlate(x, x, mode='full')
    autocorr = autocorr[x.size:]
    autocorr /= autocorr.max()

    return ax.stem(autocorr)

acorr(sampler.chain[0,:,0])
pl.show()
print ("Fig 7: Chain Autocorrelation")
```
